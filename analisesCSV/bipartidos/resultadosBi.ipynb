```python
import numpy as np 
import pandas as pd
```

```python
df_gul = pd.read_csv("resultsGulADJ_sumario.csv")
"""
df_gen_a100_b100 = pd.read_csv("results_GA_Final_a100_b100.csv")
df_gen_a100_b500 = pd.read_csv("results_GA_Final_a100_b500.csv")
df_gen_a100_b1000 = pd.read_csv("results_GA_Final_a100_b1000.csv")
df_gen_a500_b500 = pd.read_csv("results_GA_Final_a500_b500.csv")
df_gen_a500_b1000 = pd.read_csv("results_GA_Final_a500_b1000.csv")
"""

print(df_gul.columns.tolist())

"""
print(df_gen_a100_b100.columns.tolist())
print(df_gen_a100_b500.columns.tolist())
print(df_gen_a100_b1000.columns.tolist())
print(df_gen_a500_b500.columns.tolist())
print(df_gen_a500_b1000.columns.tolist())
"""
```

```python
gen_files = [
    "results_GA_Final_a100_b100.csv",
    "results_GA_Final_a100_b500.csv",
    "results_GA_Final_a100_b1000.csv",
    "results_GA_Final_a500_b500.csv",
    "results_GA_Final_a500_b1000.csv"
]

# 2. Load and combine all GA files into one DataFrame
df_gen_all = pd.concat([pd.read_csv(f) for f in gen_files], ignore_index=True)
print(df_gen_all.columns.tolist())
```

```python
# merge dos dataframes para observar resultados em número de cores e tempo do método
df_comparacao = pd.merge(df_gul, df_gen_all, on="instancia")

# mean_chi é do genético e melhor_chi é do guloso
df_comparacao['diff_chi'] = df_comparacao['mean_chi'] - df_comparacao['melhor_chi']

# mean_time é do genético e tempo_melhor é do guloso
df_comparacao['diff_tempo'] = df_comparacao['mean_time'] - df_comparacao['tempo_melhor']

# porcentagem de melhora em termos de cores para o genético
df_comparacao['pct_melhora_chi'] = (df_comparacao['diff_chi'] / df_comparacao['melhor_chi']) * 100

# "desaceleração" com o uso do genético
df_comparacao['slowdown_tempo'] = df_comparacao['mean_time'] / df_comparacao['tempo_melhor']
```

```python
estatisticas_chi = df_comparacao[['mean_chi', 'melhor_chi', 'diff_chi', 'pct_melhora_chi']].describe()
estatisticas_tempo = df_comparacao[['mean_time', 'tempo_melhor', 'diff_tempo']].describe()

print("--- Estatísticas de Cores (Chi) ---")
print(estatisticas_chi)
print("\n--- Estatísticas de Tempo ---")
print(estatisticas_tempo)

vitorias_gen = (df_comparacao['mean_chi'] < df_comparacao['melhor_chi']).sum()
empates = (df_comparacao['mean_chi'] == df_comparacao['melhor_chi']).sum()
vitorias_gul = (df_comparacao['mean_chi'] > df_comparacao['melhor_chi']).sum()

print(f"\nResumo de Performance (Cores):")
print(f"Genético melhor: {vitorias_gen}")
print(f"Empates: {empates}")
print(f"Guloso melhor: {vitorias_gul}")
```

```python
# criação do data frame com todas as instâncias já analisadas de grafos bipartidos
BRKGA_files = [
    "results_BRKGA_a100_b100.csv",
    "results_BRKGA_a100_b500.csv",
    "results_BRKGA_a100_b1000.csv",
    "results_BRKGA_a500_b500.csv",
    "results_BRKGA_a500_b1000.csv"
]

df_BRKGA = pd.concat([pd.read_csv(f) for f in BRKGA_files], ignore_index=True)
print(df_BRKGA.columns.tolist())
```

```python
# comparação entre resultados do BRKGA e dos baselines gulosos
# merge dos dataframes para observar resultados em número de cores e tempo do método
df_comp_BRKGA_gul = pd.merge(df_gul, df_BRKGA, on="instancia")

# mean_chi é do genético e melhor_chi é do guloso
df_comp_BRKGA_gul['diff_chi'] = df_comp_BRKGA_gul['mean_chi'] - df_comp_BRKGA_gul['melhor_chi']

# mean_time é do genético e tempo_melhor é do guloso
df_comp_BRKGA_gul['diff_tempo'] = df_comp_BRKGA_gul['mean_time'] - df_comp_BRKGA_gul['tempo_melhor']

# porcentagem de melhora em termos de cores para o genético - melhorar esse stat (não faz muito sentido)
df_comp_BRKGA_gul['pct_melhora_chi'] = (df_comp_BRKGA_gul['diff_chi'] / df_comp_BRKGA_gul['melhor_chi']) * 100

# "desaceleração" com o uso do genético
df_comp_BRKGA_gul['slowdown_tempo'] = df_comp_BRKGA_gul['mean_time'] / df_comp_BRKGA_gul['tempo_melhor']
```

```python
# coleta de estatísticas para resultados de BRKGA e gulosos
stats_chi_BRKGA_gul = df_comp_BRKGA_gul[['mean_chi', 'melhor_chi', 'diff_chi', 'pct_melhora_chi']].describe()
stats_tempo_BRKGA_gul = df_comp_BRKGA_gul[['mean_time', 'tempo_melhor', 'diff_tempo']].describe()

print("--- Estatísticas de Cores (Chi) ---")
print(stats_chi_BRKGA_gul)
print("\n--- Estatísticas de Tempo ---")
print(stats_tempo_BRKGA_gul)

vitorias_BRKGA = (df_comp_BRKGA_gul['mean_chi'] < df_comp_BRKGA_gul['melhor_chi']).sum()
empates_BRKGA_gul = (df_comp_BRKGA_gul['mean_chi'] == df_comp_BRKGA_gul['melhor_chi']).sum()
vitorias_GUL = (df_comp_BRKGA_gul['mean_chi'] > df_comp_BRKGA_gul['melhor_chi']).sum()

print(f"\nResumo de Performance (Cores):")
print(f"BRKGA melhor: {vitorias_BRKGA}")
print(f"Empates: {empates_BRKGA_gul}")
print(f"Guloso melhor: {vitorias_GUL}")
```

```python
# comparação dos resultados do BRKGA e do genético 

# df merged para comparar resultados do BRKGA e do genético
df_comp_BRKGA_gen = pd.merge(df_gen_all, df_BRKGA, on="instancia", suffixes=('_gen', '_BRKGA'))

# mean_chi é do BRKGA e melhor_chi é do guloso
df_comp_BRKGA_gen['diff_chi'] = df_BRKGA['mean_chi'] - df_gen_all['mean_chi']

# mean_time é do BRKGA e tempo_melhor é do guloso
df_comp_BRKGA_gen['diff_tempo'] = df_BRKGA['mean_time'] - df_gen_all['mean_time']

# porcentagem de melhora em termos de cores para o BRKGA
df_comp_BRKGA_gen['pct_melhora_chi'] = (df_comp_BRKGA_gen['diff_chi'] / df_gen_all['mean_chi']) * 100

# "desaceleração" com o uso do BRKGA
df_comp_BRKGA_gen['slowdown_tempo'] = df_BRKGA['mean_time'] / df_gen_all['mean_time']
```

```python
stats_chi_BRKGA_gen = df_comp_BRKGA_gen[['mean_chi_gen', 'mean_chi_BRKGA', 'diff_chi', 'pct_melhora_chi']].describe()
stats_tempo_BRKGA_gen = df_comp_BRKGA_gen[['mean_time_gen', 'mean_time_BRKGA', 'diff_tempo']].describe()

print("--- Estatísticas de Cores (Chi) ---")
print(stats_chi_BRKGA_gen)
print("\n--- Estatísticas de Tempo ---")
print(stats_tempo_BRKGA_gen)

vitorias_BRKGA = (df_comp_BRKGA_gen['mean_chi_BRKGA'] < df_comp_BRKGA_gen['mean_chi_gen']).sum()
empates_BRKGA_GEN = (df_comp_BRKGA_gen['mean_chi_BRKGA'] == df_comp_BRKGA_gen['mean_chi_gen']).sum()
vitorias_GEN = (df_comp_BRKGA_gen['mean_chi_BRKGA'] > df_comp_BRKGA_gen['mean_chi_gen']).sum()

print(f"\nResumo de Performance (Cores):")
print(f"BRKGA melhor: {vitorias_BRKGA}")
print(f"Empates: {empates_BRKGA_GEN}")
print(f"Genético melhor: {vitorias_GEN}")
```
